```python
import itertools

# Define graph
nodes = [0, 1, 2, 3]
edges = [(0, 1), (0, 2), (1, 2), (1, 3)]

def cut_value(bitstring, edges):
    value = 0
    for i, j in edges:
        if bitstring[i] != bitstring[j]:
            value += 1
    return value

# Enumerate all cuts
results = []
for bits in itertools.product([0, 1], repeat=len(nodes)):
    value = cut_value(bits, edges)
    results.append((bits, value))

# Sort by cut value
results.sort(key=lambda x: x[1], reverse=True)

# Display best cuts
best_value = results[0][1]
best_cuts = [r for r in results if r[1] == best_value]

best_value, best_cuts
```
