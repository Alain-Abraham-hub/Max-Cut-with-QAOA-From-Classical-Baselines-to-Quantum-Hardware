```python
from qiskit import QuantumCircuit
from qiskit.circuit import Parameter
import numpy as np

edges = [(0,1), (0,2), (1,2), (1,3)]
n_qubits = 4
```

```python
qc = QuantumCircuit(n_qubits)

for i in range(n_qubits):
    qc.h(i)

qc.draw()
```

```python
gamma = Parameter('γ')

for i, j in edges:
    qc.cx(i, j)
    qc.rz(2 * gamma, j)
    qc.cx(i, j)
```

```python
beta = Parameter('β')

for i in range(n_qubits):
    qc.rx(2 * beta, i)
```

```python
qc.measure_all()
qc.draw()
```
